```python
#Feature engineering

from pandas.core.describe import DataFrameDescriber
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Perform feature scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['OrigUPB', 'OrigLoanTerm']])

df['Scaled UPB'] = scaled_features[:, 0]
df['Scaled Loan Term'] = scaled_features[:, 1]

# Extract new features
df['Loan-to-Value Ratio'] = df['OrigUPB'] / df['OrigInterestRate']
df['Monthly Payment'] = df['OrigInterestRate'] / df['OrigLoanTerm']

df['Scaled UPB'] = scaled_features[:, 0]
df['Scaled Loan Term'] = scaled_features[:, 1]

# Print the updated dataset
print(df.head())
```

```python
# Perform feature scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['FirstPaymentDate', 'Scaled UPB', 'Monthly Payment']])

df['Scaled MIP'] = scaled_features[:, 0]
df['Scaled DTI'] = scaled_features[:, 1]
df['Scaled OCLTV'] = scaled_features[:, 2]
df['Scaled Units'] = scaled_features[:, 2]

# Drop original columns
df.drop(['Scaled UPB', 'MSA', 'CreditScore'], axis=1, inplace=True)
```

```python
ltv_ranges = [(0, 80), (80, 90), (90, 100)]
ltv_labels = ['Low', 'Medium', 'High']

# Create a new column 'LTV Category' based on the LTV ranges
df['LTV Category'] = pd.cut(df['LTV'], bins=[range[0] - 1 for range in ltv_ranges] + [ltv_ranges[-1][1]], labels=ltv_labels)

# Print the updated dataset
print(df.head())
```

```python
# Convert 'FirstTimeHomeBuyer' column to numerical values
label_encoder = LabelEncoder()
df['FirstTimeHomebuyer'] = label_encoder.fit_transform(df['FirstTimeHomebuyer'])
df['YearsInRepayment'] = df['MonthsInRepayment'] / 12
df['YearsInRepayment'].head()
```

```python
df['MonthsInRepayment'].head()
```

```python
range_limits = [0, 4, 8, 12, 16, np.inf]
range_labels = ['Range 1', 'Range 2', 'Range 3', 'Range 4', 'Range 5']

# Create a new column 'RepayRange' based on the ranges
df['RepayRange'] = pd.cut(df['YearsInRepayment'], bins=range_limits, labels=range_labels, right=False)
df.head()
```

```python
df['LTV_bin'] = pd.cut(df['LTV'], bins=[0, 579, 669, 739, 799, float('inf')],
                                 labels=['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'])
df['LTV_diff'] = df['LTV'] - df.groupby('LoanPurpose')['LTV'].transform('mean')
df['LTV_binary'] = df['LTV'].apply(lambda x: 1 if x > 800 else 0)
df['LTV_bin'] = df['LTV_bin'].cat.add_categories('Missing').fillna('Missing')
df['LTV_bin'].fillna(0, inplace=True)


df['RemainingMaturity'] = pd.to_datetime(df['MaturityDate'], errors='coerce').sub(pd.to_datetime('today')).dt.days
df['RemainingMaturity'].fillna(0, inplace=True)


# MaturityDate
df['MaturityDate_binary'] = df['MaturityDate'].apply(lambda x: 1 if x > 0 else 0)



df['Channel_group'] = pd.to_numeric(df['Channel'], errors='coerce')

df['Channel_group'] = df['Channel_group'].apply(lambda x: 1 if x > 0 else 0)

df['Channel_group'].fillna(0, inplace=True)


df['Occupancy'] = df['Occupancy'].map({'O': 1, 'I': 0})
df['Occupancy'].fillna(0, inplace=True)


df['LTV_bin'] = pd.cut(df['LTV'], bins=[0, 60, 70, 80, 90, 100, float('inf')],
                           labels=['<60', '60-69', '70-79', '80-89', '90-100', '>100'])
df['LTV_diff'] = df['LTV'] - df.groupby('LoanPurpose')['LTV'].transform('mean')
df['LTV_bin'] = df['LTV_bin'].cat.add_categories('Missing').fillna('Missing')



df['OrigUPB_bin'] = pd.cut(df['OrigUPB'], bins=[0, 20, 30, 40, 50, float('inf')],
                         labels=['<20', '20-29', '30-39', '40-49', '>50'])
df['OrigUPB_diff'] = df['OrigUPB'] - df.groupby('LoanPurpose')['OrigUPB'].transform('mean')
df['OrigUPB_bin'] = df['OrigUPB_bin'].cat.add_categories('Missing').fillna('Missing')


df['OrigUPB_bin'] = pd.cut(df['OrigUPB'], bins=[0, 100000, 200000, 300000, float('inf')],
                             labels=['<100k', '100k-199k', '200k-299k', '>300k'])
df['OrigUPB_bin'] = df['OrigUPB_bin'].cat.add_categories('Missing').fillna('Missing')


# LTV
df['LTV_bin'] = pd.cut(df['LTV'], bins=[0, 60, 70, 80, 90, 100, float('inf')],
                         labels=['<60', '60-69', '70-79', '80-89', '90-100', '>100'])
df['LTV_diff'] = df['LTV'] - df.groupby('LoanPurpose')['LTV'].transform('mean')
df['LTV_bin'] = df['LTV_bin'].cat.add_categories('Missing').fillna('Missing')


df['OrigInterestRate_bin'] = pd.cut(df['OrigInterestRate'], bins=[0, 3, 4, 5, float('inf')],
                                      labels=['<3', '3-3.99', '4-4.99', '>=5'])



if 'Channel' in df.columns:
    df = pd.get_dummies(df, columns=['Channel'], prefix='Channel')



df['PPM_binary'] = df['PPM'].apply(lambda x: 1 if x == 'Y' else 0)
df['PPM_binary'].fillna(0, inplace=True)


#if 'LoanPurpose' in df.columns:
    #df = pd.get_dummies(df, columns=['LoanPurpose'], prefix='LoanPurpose')


df['OrigLoanTerm_bin'] = pd.cut(df['OrigLoanTerm'], bins=[0, 300, 400, float('inf')],
                                  labels=['Short', 'Medium', 'Long'])


df['NumBorrowers'] = pd.to_numeric(df['NumBorrowers'], errors='coerce')

# Replace any NaN values with 0
df['NumBorrowers'].fillna(0, inplace=True)


df['NumBorrowers'] = df['NumBorrowers'].apply(lambda x: 1 if x > 1 else 0)


df.columns
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#columns for PCA
columns_for_pca = ['OrigUPB_diff','LTV_diff']


# Standardize the selected columns
scaler = StandardScaler()
df_selected = df[columns_for_pca]
df_selected_scaled = scaler.fit_transform(df_selected)

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_selected_scaled)

# results
explained_variance = pca.explained_variance_ratio_
principal_components = pd.DataFrame(
    data=pca.components_,
    columns=columns_for_pca,
    index=['PC1', 'PC2']
)

# explained variance ratio
print("Explained variance ratio:", explained_variance)

# principal components
print("Principal components:")
print(principal_components)
```
